```python
def _progress(a,b):
    sys.stdout.write('\r')
    p_a, p_b = (a/b)*30, (1-(a/b))*30
    p_ai = int(p_a)
    p_bi = int(p_b) + 1
    
    sys.stdout.write('*'*p_ai+ '-' * p_bi)  
    sys.stdout.flush()
    
    
    
    
    
_progress(7,11)
_progress(9,11)
```

```python
for i in range(151):
    _progress(i,150)
    sleep(.05)
```
